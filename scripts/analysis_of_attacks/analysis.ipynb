```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("AmitX.csv")
```

```python


df
```

```python
x = np.arange(50)
y1 = df["Original Malware"]
y2 = df["First Submission of Attack"]
y3 = df["Baseline Attack"]
# y4 = df["X"]
# y5 = df["evade_noimg_withgdi_addedfunc2_results"]
y6 = df["image strategy"]
y7 = df["no image with encoding"]
# y8 = df["addfunc_no_img_without_gdi"]
# y9 = df["AddedFunc2_Img"]

plt.figure(figsize=(38,14), dpi=120)

plt.plot(x, y1, '-.', label ='Original Malware', linewidth = 2.5)
plt.plot(x, y2, '-.', label ='First Submission of Attack', linewidth = 2.5)
plt.plot(x, y3, '-.', label ='Baseline Attack', linewidth = 2.5)
# plt.plot(x, y4, '-.', label ='Final Submission of Attack', linewidth = 2.5)
plt.plot(x, y6, '-.', label ='image strategy', linewidth = 2.5)
plt.plot(x, y7, '-.', label ='no image with encoding', linewidth = 2.5)

# plt.plot(x, y7, label ='addfunc_no_img_with_gdi32')
# plt.plot(x, y8, label ='addfunc_no_img_without_gdi')
# plt.plot(x, y9, label ='AddedFunc2_Image')

plt.xticks(np.arange(0, 50, 1.0))
plt.yticks(np.arange(0, 100, 10.0))
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.legend()
plt.title('Analysis of Malware Strategies for Model - Amit')
plt.savefig('amit_IamConfusion1.png', dpi=120)
plt.show()
```
